```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import rcParams
rcParams['figure.figsize']=10,6
```

```python
sale=pd.read_excel('Case Data 0322.xlsx',sheet_name='Sale',skiprows=1)
```

```python
sale.head()
```

```python
sale.drop(['Type','Location'],inplace=True,axis=1)
```

```python
s=sale.copy()
```

```python
s.head()
```

```python
s['SKU'] = s['New SKU'].str[14:]
```

```python
dummies = pd.get_dummies(s['Brand'])
dum=pd.get_dummies(s['Upstream Contact'])
```

```python
dumm=pd.get_dummies(s['New SKU'])
```

```python
w=pd.concat([s,dummies,dum,dumm],axis=1)
```

```python
w.head()
```

```python
w.drop(['New SKU','Upstream Contact','Brand','SKU'],inplace=True,axis=1)
```

```python
w.head()
```

```python
w.columns
```

```python
#Elbow Method to find optimal no of clusters (k)
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

cluster_range = range (1,10)
cluster_errors=[]
for num_clusters in cluster_range:
    clusters = KMeans(num_clusters)
    clusters.fit(w)
    cluster_errors.append(clusters.inertia_) #clusters.inertia_ calculates wss (within sum of squares)
plt.figure(figsize=(6,4))
plt.plot(cluster_range, cluster_errors,marker="o" );
plt.title("Elbow Diagram")
```

```python
#K-means Clustering
#Based on elbow diagram, we have choosen optimal cluster no as 3
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=42 )
clusters_new.fit( w )
w["clusterid"] = clusters_new.labels_
```

```python
#Finding Cluster Centroid
centers = np.array(clusters_new.cluster_centers_)
centers=pd.DataFrame(centers)
centers=centers.T
centers=centers.rename(columns={0: "C1", 1: "C2",2: "C3"})
```

```python
Trends_clustered=pd.merge(sale, centers, left_index=True, right_index=True)
Trends_clustered.head()
```

```python
q=w['clusterid']
q.head()
```

```python
Trends_clustered=pd.concat([Trends_clustered,q],axis=1)
```

```python
#Cluster 1
print("Cluster 1 SKU: ")
print(Trends_clustered[Trends_clustered['clusterid'] == 0][['New SKU']].values)

#Cluster 2
print("Cluster 2 SKU: ")
print(Trends_clustered[Trends_clustered['clusterid'] == 1][['New SKU']].values)

#Cluster 3
print("Cluster 3 SKU: ")
print(Trends_clustered[Trends_clustered['clusterid'] == 2][['New SKU']].values)
```

```python
#Cluster 1 
Trends_clustered.sort_values('C1', ascending=False).head(10)
```

```python
#Cluster 2 
Trends_clustered.sort_values('C2', ascending=False).head(10)
```

```python
#Cluster 3
Trends_clustered.sort_values('C3', ascending=False).head(10)
```

```python
cluster1=Trends_clustered[Trends_clustered['clusterid'] == 0]
cluster2=Trends_clustered[Trends_clustered['clusterid'] == 1]
cluster3=Trends_clustered[Trends_clustered['clusterid'] == 2]
```
